 ### DataSet description
Description of columns in the files:
- Rk - Row Index/rank 
- Player - Player Name
- Nation - Nationality Of the Player
- Pos - Position Of The Player
- Squad - Team Name
- Age - Years old
- Born - Date of Birth
- Mp - Matches Played
- Starts - Match Started 
- Min - Minutes Played Total 
- 90s - Average Minutes Played in a Match
- Gls - Goals Scored
- Ast- Assist Given 
- GlsAst - Goals + Assist
- G-Pk - Goals - Penalty Goals 
- Pk - Penalty Goals
- PKatt - Penalty Attempted
- CrdY - Yellow card received
- CrdR - Red Card received 
- xG - Expected Goals 
- npxG - Non-penalty Expected Goals
- xAG - Expected Assisted Goals 
- npxG+xAG - Non-penalty Expected Goals + Expected Assisted Goals 
- PrgC - Progressive Carries 
- PrgP - Progressive Passes
- PrgR - Progressive passes received 
- Gls90 - Goals per Match/90 Minutes 
- Ast.1 - Assist per Match/90 Minutes 
- G+A - Goals + Assist per Match/90 Minutes 
- G-PK.1 - Goals Minus Penalty Goals Per 90 Minutes
- xG.1 - Expected Goals per 90 Minutes
- xAG.1 - Expected Assists Per 90 Minutes
- xG+xAG - Expected Goals + Expected Assist per 90 Minutes
- npxG.1 - Non-Penalty Expected Goals Per 90 Minutes 
- npxG+xAG.1 - Non-Penalty expected goals per 90 Minutes  + expected Goals per 90 Minutes 
- Matches - Matches Details Link

#### Feature Engineering:
- Creating a new categorical column 'age-type'
  

```python
def type1(x):    
    if x>=31:
        return 'Experienced'
    elif x>20 and x<=30:
        return 'Young'
    elif x<=20:
        return 'Teenage'    
    
```

## Analysis & Visualization on The Dataset


### Visualization of Goal Conversion & Penalty Conversion

#### Total Goals In Epl 2022-23

```python
Total_Goals=epl_df['Gls'].sum()
print(Total_Goals)
```

*The Epl 2022-23 Season Had* **1039** *Goals In Total* 

#### Non-Peanlty Goals

```python
Total_NonPenaltyGoals=epl_df['G-PK'].sum()
print(Total_NonPenaltyGoals)
```

**965** *Non-Peanlty Goals*

#### Total Penalty Goals

```python
Total_PenaltyGoals=epl_df['PK'].sum()
print(Total_PenaltyGoals)
```

 **74** Peanlty Goals

#### Total Penalty Attemptes

```python
Total_penaltyAttempt=epl_df['PKatt'].sum()
print(Total_penaltyAttempt)
```

**99** Penalty Attemptes

#### Expected Goals

```python
Total_Xg=epl_df['xG'].sum()
print(Total_Xg)
```

**1103.9** *Expected Goals*

```python
data=[Total_Goals,Total_Xg-Total_Goals]
c=['Scored','Missed']
sns.set_style("white")
plt.subplot(1,2,1)
plt.pie(data,autopct='%2.1f%%',shadow=True,explode=[0.3,0],colors=sns.color_palette("light:b"),labels=c)
plt.title('Goals Conversion Rate')


data1=[Total_PenaltyGoals,Total_penaltyAttempt-Total_PenaltyGoals]
plt.subplot(1,2,2)
plt.pie(data1,autopct='%2.1f%%',shadow=True,explode=[0.1,0],colors=sns.color_palette("flare"),labels=c)
plt.title('Penalty Goals Conversion Rate')
plt.tight_layout()
plt.show()
```

#### Analysis & Visualization on Age

```python
Under20=epl_df[epl_df['Age_type']=='Teenage'].groupby('Squad').agg(sz=('Squad','size')).reset_index()
Under20
```

```python
age21_30=epl_df[epl_df['Age_type']=='Young'].groupby('Squad').agg(sz=('Squad','size')).reset_index()
age21_30
```

```python
age31_above=epl_df[epl_df['Age_type']=='Experienced'].groupby('Squad').agg(sz=('Squad','size')).reset_index()
new_row = pd.DataFrame({'Squad': ['Aresnal'], 'sz': [0]})
age31_above= pd.concat([age31_above,new_row], ignore_index=True)
age31_above=age31_above.sort_values(by='Squad')
age31_above
```

#### Visualization of diffrent age group

```python
plt.figure(figsize=(14,10))
plt.subplot(3,1,1)
plt.bar(Under20['Squad'],Under20['sz'])
plt.xticks(rotation=60)
plt.ylim(0,26)
plt.title('Player Aged Under or 20')

plt.subplot(3,1,2)
plt.bar(age21_30['Squad'],age21_30['sz'])
plt.xticks(rotation=60)
plt.ylim(0,26)
plt.title('Players Aged Between 21-30')

plt.subplot(3,1,3)
plt.bar(age31_above['Squad'],age31_above['sz'])
plt.xticks(rotation=60)
plt.ylim(0,26)
plt.title('Players Aged Between 31 And Above')

plt.tight_layout()
#plt.legend(lbl,loc=4,framealpha=1)
plt.show()
```
